```python
class Triangle:

    def __init__(self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    def perimeter(self):
        if self.side3 < (self.side1+self.side2) and self.side2 < (self.side1+self.side3) and self.side1 < (self.side2+self.side3):
            return self.side1 + self.side2 + self.side3
            
        else:
            print("Given sides do not form a triangle...")
    
    def area(self):
        s = self.perimeter()/2
        a = (s*(s-self.side1)*(s-self.side2)*(s-self.side3))**0.5
        print(f"Area = {a:.2f}")

```

```python
t = Triangle(5,3,2)
```

```python
t.perimeter()
```

```python
t1 = Triangle(12,13,14)
```

```python
t1.perimeter()
```

```python
t1.area()
```
