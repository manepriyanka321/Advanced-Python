### How does a triangle form
1. Triangle inewuality sum of any tow sides should be greater than thied
2. Perimeter is sum of all sides
3. Area is calculates by Herons formula


```python
from dataclasses import dataclass
```

```python
@dataclass
class Triangle:
    a = int | float
    b = int | float
    c = int | float

    def is_triangle(self):
        if ((self.a + self.b > self.c) and
            (self.b + self.c > self.a) and
            (self.c + self.a > self.b)):
            return True
        else: 
            return False
        
    def perimeter(self):
        if self.is_triangle():
            p = self.a + self.b + self.c
            return p
        
        else:
            raise ValueError(f"Given sides {self.a}, {self.b}, {self.c} do not form a triangle")
    
    def area(self):
        s = self.perimeter()/2
        a = (s*(s-self.a)*(s-self.b)*(s-self.c))**0.5
        return a
```

```python
@dataclass
class Triangle:
    a: int | float
    b: int | float
    c: int | float

    def is_triangle(self):
        if ((self.a + self.b > self.c) and 
            (self.b + self.c > self.a) and 
            (self.c + self.a > self.b)):
            return True
        else:
            return False
        
    def perimeter(self):
        if self.is_triangle():
            p = self.a + self.b + self.c
            return p
        else:
            raise ValueError(f"Given sides {self.a}, {self.b}, {self.c} do not form a triangle")
    
    def area(self):
        s = self.perimeter() / 2
        a = (s * (s - self.a) * (s - self.b) * (s - self.c))**(1 / 2)
        return a
```

### Calculate area and perimeter for different shapes

```python
@dataclass
class Rectangle:
    height: int | float
    width : int | float

    def perimeter(self):
        return 2*(self.height + self.width)
    
    def area(self):
        return self.width * self.height
```

```python
from math import pi
```

```python
pi
```

```python
@dataclass
class Circle:
    radius : int | float

    def perimeter(self):
        return 2*pi*self.radius
    def area(self):
        return pi*self.radius**2
```

```python
shapes = [
    Triangle(3,4,5),
    Circle(7),
    Rectangle(10,20),
    Circle(20),
    Triangle(5,7,9),
    Rectangle(15, 23)
]
```

```python
for i in shapes:
    print(i, type(i))
    print(f"Perimeter : {i.perimeter():.2f}")
    print(f"Area : {i.area():.2f}")
    print("\n","= * "*25, "=","\n")
```
